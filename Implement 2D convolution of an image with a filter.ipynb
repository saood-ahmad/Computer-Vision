#### Prog.  Implement 2D convolution of an image with a filter. Show the output image.
___

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('india.png',0)
row,col=img.shape
# box filter kernel 
kernel=np.array([[0.11111111,0.111111111,0.11111111],
                [0.11111111,0.11111111,0.11111111],
                [0.11111111,0.11111111,0.11111111]])
box_img = np.zeros((row,col))
for i in range(1,row-1):
    for j in range(1,col-1):
        m = np.array([[img[i-1][j-1],img[i-1][j],img[i-1][j+1]],
                     [img[i][j-1],img[i][j],img[i][j+1]],
                     [img[i+1][j-1],img[i+1][j],img[i+1][j+1]]])
        mk = m*kernel
        value = np.sum(mk) 
        box_img[i][j] = value
        #print(value)
f, axarr = plt.subplots(1,2, figsize=(15, 15))

axarr[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axarr[1].imshow(box_img, cmap='gray', vmin=0, vmax=255)
axarr[0].title.set_text('Original Image')
axarr[1].title.set_text('After Box filter')
```
