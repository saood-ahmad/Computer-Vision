#### Prog.  Implement Histogram equalization of an image. Compare the original image and histogram equalized image.
___

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
gray_img = cv2.imread('low.jpg',0)
row,col = gray_img.shape
gray_scale = np.zeros(256)   # grey Level
# no. of pixel
for i in range(row):
    for j in range(col):
        gray_scale[gray_img[i][j]] += 1
        
#print(gray_scale)
# adding values
for k in range(len(gray_scale)-1):
    gray_scale[k+1] = gray_scale[k] + gray_scale[k+1]
# divide by total no. of pixel
gray_norm = gray_scale/gray_scale[-1]
# multiply by no.of gray level
gray_norm_1 = gray_norm*255
cont_img = np.zeros(gray_img.shape)
for i in range(row):
    for j in range(col):
        cont_img[i][j] = gray_norm_1[gray_img[i][j]]

f, axarr = plt.subplots(1,2, figsize=(15, 15))

axarr[0].imshow(gray_img, cmap='gray', vmin=0, vmax=255)
axarr[1].imshow(cont_img, cmap='gray', vmin=0, vmax=255)
axarr[0].title.set_text('Original Image')
axarr[1].title.set_text('After Histogram equalization ')
```
