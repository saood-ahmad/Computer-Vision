```python
___
#### Prog. Implement to get the 2D DFT of an image. Plot the magnitude and phase spectrum.
___

import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('work.jpg',0)

dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
# find magintude spectrum
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

img = cv2.imread('work.jpg',0)
dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)

# Phase Spectrum
phase_spectrum = np.angle(dft_shift)


f, axarr = plt.subplots(1,3, figsize=(15, 15))

axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(magnitude_spectrum, cmap='gray')
axarr[2].imshow(phase_spectrum, cmap='gray')
axarr[0].title.set_text('Original Image')
axarr[1].title.set_text('Magnitude Spectrum')
axarr[2].title.set_text('Phase Spectrum')
```
