#### Prog. 1) To generate square wave using Fourier series using N terms.

### square wave using Fourier series using 10 terms

```python
#import library
import numpy as np
from matplotlib import pyplot as plt
time        = np.arange(0, 2*np.pi, 0.01);

# Amplitude of the sine wave is sine of a variable like time

amplitude   = np.sin(2*np.pi*time)
a = 3
n = 10
for i in range(3, n+2, 2):

# Plot a sine wave using time and amplitude obtained for the sine wave
    amplitude_temp= np.sin(2*np.pi*i*time)/i
    amplitude += amplitude_temp
    #plt.plot(time,ampt)

#ampt= amplitude + amplitude2 + amplitude3 + amplitude4 +amplitude1
plt.plot(time,amplitude, )
plt.title('Square wave using Fourier series using 10 terms')
```
